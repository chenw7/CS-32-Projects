```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(-10,10+0.2,0.2)

def func(t):
  return np.exp(-t)*np.cos(2*np.pi*t)

y = func(t)
plt.plot(t,y)
plt.xlabel("t")
plt.ylabel("f(t)")
plt.grid()
plt.title("6a. e^(-t)*cos(2*pi*t)")
```

```python
t = np.arange(-5,5+0.1,0.1)

def relu(t):
  return np.maximum(0,t)

plt.plot(t,relu(t))
plt.title("6b. ReLU function")
plt.grid()
plt.xlabel("t")
plt.ylabel("x(t)")
```

```python
def even(t,f):
  return 0.5*(f(t)+f(-t))

def odd(t,f):
  return 0.5*(f(t)-f(-t))

t = np.arange(-5,5+0.1,0.1)

y_even = [even(tt,relu) for tt in t]
y_odd = [odd(tt,relu) for tt in t]

fig, ax = plt.subplots(1,3)
fig.tight_layout()

ax[0].plot(t,y_even)
ax[0].set_xlabel('t')
ax[0].set_ylabel('function')
ax[0].set_title('6c. ReLU Even')
ax[0].grid()

ax[1].plot(t,y_odd)
ax[1].set_xlabel('t')
ax[1].set_ylabel('function')
ax[1].set_title('ReLU Odd')
ax[1].grid()

ax[2].plot(t,y_even,label='even')
ax[2].plot(t,y_odd,label='odd')
ax[2].set_xlabel('t')
ax[2].set_ylabel('function')
ax[2].set_title('Both Components of ReLU')
ax[2].legend()
ax[2].grid()

plt.show()
```
